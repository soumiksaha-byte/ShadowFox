```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/Sample - Superstore.csv', encoding='ISO-8859-1')  # or encoding='latin1'
df.head()

```

```python
print(df.info())
print(df.describe())
print(df.isnull().sum())
```

```python
total_sales = df['Sales'].sum()
total_profit = df['Profit'].sum()
average_discount = df['Discount'].mean()
print(f"Total Sales: ${total_sales:.2f}")
print(f"Total Profit: ${total_profit:.2f}")
print(f"Average Discount: {average_discount:.2f}")
```

```python
segment_profit = df.groupby('Segment')['Profit'].sum().sort_values(ascending=False)
segment_profit.plot(kind='bar', title='Profit by Segment')
plt.ylabel("Total Profit")
plt.show()
```

```python
region_profit = df.groupby('Region')['Profit'].sum().sort_values()
region_profit.plot(kind='barh', color='orange', title='Profit by Region')
plt.xlabel("Profit")
plt.show()
```

```python
category_profit = df.groupby('Category')['Profit'].sum().sort_values(ascending=False)
subcategory_profit = df.groupby('Sub-Category')['Profit'].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 5))
category_profit.plot(kind='bar', title='Profit by Category')
plt.ylabel("Profit")
plt.show()

plt.figure(figsize=(12, 6))
subcategory_profit.plot(kind='bar', title='Profit by Sub-Category', color='green')
plt.ylabel("Profit")
plt.show()
```

```python
category_profit = df.groupby('Category')['Profit'].sum().sort_values(ascending=False)
subcategory_profit = df.groupby('Sub-Category')['Profit'].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 5))
category_profit.plot(kind='bar', title='Profit by Category')
plt.ylabel("Profit")
plt.show()

plt.figure(figsize=(12, 6))
subcategory_profit.plot(kind='bar', title='Profit by Sub-Category', color='green')
plt.ylabel("Profit")
plt.show()
```

```python
sns.scatterplot(data=df, x='Sales', y='Profit', hue='Category')
plt.title("Sales vs Profit")
plt.show()
```

```python
sns.lineplot(data=df, x='Discount', y='Profit')
plt.title("Discount vs Profit Trend")
plt.show()
```
